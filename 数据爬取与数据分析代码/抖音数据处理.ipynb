```python
typefile = open('H:\\NJU Documents\\抖音数据分析ppt\\数据分析\\淘宝商品类目.txt')
typeDictionaryList = []
line = typefile.readline()
lineList = line.split('；')
typeDictionaryList.append(lineList)
while line:
    line = typefile.readline()
    lineList = line.split('；')
    typeDictionaryList.append(lineList)
typeDictionaryList = typeDictionaryList[0:len(typeDictionaryList) - 1]
bgmFrame = bgmFrame.reset_index(drop=True)
fenleiFrame = pd.DataFrame(columns=['title', 'type', 'similarity', 'duration', 'sale', 'date', 'bgm'])
for i in range(len(fenciFrame)):
    fenleiFrame.loc[i, 'title'] = bgmFrame.iloc[i]['title']
    fenleiFrame.loc[i, 'duration'] = bgmFrame.iloc[i]['duration']
    fenleiFrame.loc[i, 'sale'] = bgmFrame.iloc[i]['sale']
    fenleiFrame.loc[i, 'date'] = bgmFrame.iloc[i]['date']
    fenleiFrame.loc[i, 'bgm'] = bgmFrame.iloc[i]['bgm']
    sentence = bgmFrame.iloc[i]['title']
    print(sentence)
    maxSimilarList = []
    maxMatchType = []
    print(lcut(sentence))
    for j in range(len(typeDictionaryList)):
        typeList = typeDictionaryList[j]
        typename = typeDictionaryList[j][0]
        typetexts = [lcut(type) for type in typeList]
        dictionary = Dictionary(typetexts)
        num_features = len(dictionary.token2id)
        corpus = [dictionary.doc2bow(type) for type in typetexts]
        kw_vector = dictionary.doc2bow(lcut(sentence))
        tfidf = TfidfModel(corpus)
        tf_texts = tfidf[corpus]
        tf_kw = tfidf[kw_vector]
        sparse_matrix = SparseMatrixSimilarity(tf_texts, num_features)
        similarities = sparse_matrix.get_similarities(tf_kw)
        tempList = []
        tempTypeList = []
        for e, s in enumerate(similarities, 1):
            tempList.append(s)
            tempTypeList.append(typeList[e - 1])
        res = max(tempList)
        tempIndex = tempList.index(res)
        print(str(tempTypeList[tempIndex]) + ' ' + str(res))
        maxSimilarList.append(res)
        maxMatchType.append(typename)
    resSimilar = max(maxSimilarList)
    index = maxSimilarList.index(resSimilar)
    for k in range(len(maxSimilarList) - 1, -1, -1):
        if maxSimilarList[k] == resSimilar:
            index = k
            break
    resType = maxMatchType[index]
    print('> ' + str(resType) + ' ' + str(resSimilar))
    print('end')
    fenleiFrame.loc[i, 'similarity'] = resSimilar
    if resSimilar > 0.3:
        fenleiFrame.loc[i, 'type'] = resType
```
