```python
import librosa
import numpy as np
import pandas as pd
```

```python
path = r'H:\NJU Documents\抖音数据分析ppt\歌曲下载\新建文件夹\「二胡」雨碎江南.wav'

yy, sr = librosa.load(path)
onset_env = librosa.onset.onset_strength(yy, sr=sr, hop_length=512, aggregate=np.median)
tempo, _ = librosa.beat.beat_track(onset_envelope=onset_env, sr=sr)
```

```python
modelFrame = pd.read_csv(r'H:\NJU Documents\tiktok\数据分析\建模数据集Final.csv')
```

```python
modelFrame['bpm'] = ''
```

```python
datasetFrame = pd.read_csv(r'H:\NJU Documents\抖音数据分析ppt\数据分析\音乐数据库Final.csv')
```

```python
datasetFrame['bpm'] = ''
```

```python
datasetFrame
```

```python
for i in range(0,len(modelFrame)):
    path = 'S:\\抖音音频\\BGM\\新建文件夹\\' + modelFrame.iloc[i]['filename'] + '.wav'
    yy, sr = librosa.load(path)
    onset_env = librosa.onset.onset_strength(yy, sr=sr, hop_length=512, aggregate=np.median)
    tempo, _ = librosa.beat.beat_track(onset_envelope=onset_env, sr=sr)
    modelFrame.loc[i,'bpm'] = tempo
    print('已处理' + str(i + 1) + '首歌曲')
```

```python
for i in range(len(datasetFrame)):
    path = 'S:\\抖音音频\\歌曲数据库\\' + datasetFrame.iloc[i]['bgm'] + '.wav'
    yy, sr = librosa.load(path)
    onset_env = librosa.onset.onset_strength(yy, sr=sr, hop_length=512, aggregate=np.median)
    tempo, _ = librosa.beat.beat_track(onset_envelope=onset_env, sr=sr)
    datasetFrame.loc[i,'bpm'] = tempo
    print('已处理' + str(i + 1) + '首歌曲')
```

```python
modelFrame.to_csv(r'H:\NJU Documents\抖音数据分析ppt\数据分析\建模数据集Final.csv')
```
