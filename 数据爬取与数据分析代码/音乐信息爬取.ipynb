```python
from selenium import webdriver
from time import sleep
import pandas as pd
from selenium.common.exceptions import NoSuchElementException, ElementNotVisibleException
import pyperclip
```

```python
def searchSongInfo(driver, song):
    infoList = []
    try:
        driver.find_element_by_class_name('search_input__input').send_keys(song)
        driver.find_element_by_class_name('sprite').click()
        sleep(2)
        ul_songlist = driver.find_element_by_xpath('//*[@id="song_box"]/div[2]/ul[2]')
        li_js_songlist__child = ul_songlist.find_element_by_xpath('//*[@id="song_box"]/div[2]/ul[2]/li[1]')
        li_js_songlist__child.find_element_by_xpath(
            '//*[@id="song_box"]/div[2]/ul[2]/li[1]/div/div[2]/span[1]/a/span[1]').click()
        sleep(2)
        infoList.append(driver.find_element_by_class_name('js_lan').text)
        infoList.append(driver.find_element_by_class_name('js_genre').text)
        infoList.append(driver.find_element_by_class_name('js_public_time').text)
        driver.find_element_by_id('copy_link').click()
        lyrics = pyperclip.paste()
        f = open('H:\\NJU Documents\\抖音数据分析ppt\\歌词\\' + song + '.txt', mode='w', encoding='utf-8')
        f.write(lyrics)
        f.close()
    except NoSuchElementException as e:
        pass
    except ElementNotVisibleException:
        pass
    print(infoList)
    return infoList
```

```python
dataFrame = pd.read_csv(r'H:\NJU Documents\抖音数据分析ppt\data3.csv', error_bad_lines=False)
bgmFrame = dataFrame.drop_duplicates('bgm', keep='first').reset_index(drop=True)
bgmFrame['language'] = ''
bgmFrame['genre'] = ''
bgmFrame['public_time'] = ''
songList = []
for i in range(1, len(bgmFrame)):
    song = bgmFrame.iloc[i]['bgm']
    index = song.find('-')
    song = song[0:index - 1]
    songList.append(song)
driver = webdriver.Chrome(r'L:\下载\chromedev-v71.0.3573.0\chromedriver.exe')
driver.maximize_window()
driver.get("https://y.qq.com/")
sleep(10)
driver.find_element_by_class_name('popup__icon_close').click()
cnt = 0
for song in songList:
    infoList = searchSongInfo(driver, song)
    if len(infoList) == 3:
        language = infoList[0]
        genre = infoList[1]
        public_time = infoList[2]
        index1 = language.find('：')
        index2 = genre.find('：')
        index3 = public_time.find('：')
        language = language[index1 + 1:].replace(' ', '')
        genre = genre[index2 + 1:].replace(' ', '')
        public_time = public_time[index3 + 1:].replace(' ', '')
        bgmFrame.loc[cnt + 1, 'language'] = language
        bgmFrame.loc[cnt + 1, 'genre'] = genre
        bgmFrame.loc[cnt + 1, 'public_time'] = public_time
    cnt = cnt + 1
    print('完成' + str(cnt) + '首歌曲的搜索')
```
